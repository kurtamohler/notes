`geqrf` is a LAPACK function that performs QR decomposition of a matrix. [documentation](https://www.netlib.org/lapack/explore-html/d0/da1/group__geqrf_ga01418ec8a343078109cde4febb6cda7d.html#ga01418ec8a343078109cde4febb6cda7d)


It takes an input matrix $A$ of size $m \times n$ and decomposes it into two matrices, $Q$ and $R$, such that:

* $A = Q \begin{pmatrix}R \\ 0\end{pmatrix}$
* $Q$ is size $m \times m$
* $R$ is size $n \times n$
* $\begin{pmatrix}R \\ 0\end{pmatrix}$ is size $m \times n$
* $Q$ is orthonormal, so $Q^T Q = I$
* $R$ is upper triangular, so $R_{ij} = 0$ for $j < i$
